```python
# pair plot ( diag, offdiag)

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

a = sns.load_dataset('tips')
g = sns.PairGrid(a)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot)
# 1st run till line 8, then 9, then 10 to understand it properly
```

```python
# pair plot ( scatter)

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

a = sns.load_dataset('tips')
g = sns.PairGrid(a)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.scatterplot)
plt.show()
```
