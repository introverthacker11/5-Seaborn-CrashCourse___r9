```python
# catplot

import seaborn as sns
import pandas as pd

a = pd.read_csv('groupby(csvfile8).csv')
print(a)
sns.catplot( x = 'College', y = 'Age', data = a)
```

```python
# vertical plot

import seaborn as sns
import pandas as pd

a = pd.read_csv('groupby(csvfile8).csv')
print(a)
sns.catplot( y = 'College', x = 'Age', data = a)
```

```python
# jitter 

import pandas as pd
import seaborn as sns

a = pd.read_csv('groupby(csvfile8).csv')
sns.catplot( x = 'College', y = 'Age', data = a, jitter = False)
```

```python
# swarm plot (it does not support style or size)

import pandas as pd
import seaborn as sns

a = pd.read_csv('groupby(csvfile8).csv')
sns.catplot( x = 'College', y = 'Age', data = a, kind = 'swarm')
```

```python
# swarm plot (it does not support style or size)

import pandas as pd
import seaborn as sns

a = pd.read_csv('groupby(csvfile8).csv')
sns.catplot( x = 'College', y = 'Age', data = a, kind = 'swarm', hue = 'Department')
```

```python
# swarm plot (it does not support style or size)

import seaborn as sns

a = sns.load_dataset('tips')
sns.catplot( x = 'smoker', y = 'tip', data = a, kind = 'swarm', hue = 'time')
```

```python
# passing order

import seaborn as sns

a = sns.load_dataset('tips')
colors = { 'Lunch' : 'Red' , "Dinner" : 'Blue'}
sns.catplot( x = 'smoker', y = 'tip', data = a, order = (['No','Yes']), hue = 'time', palette = colors, kind = 'swarm')
```
