```python
# hue (hue emphasizes the distinction between categories through color)

import seaborn as sns

a = sns.load_dataset('tips').tail(10)
print(a)

sns.relplot( x = 'total_bill', y = 'tip', data = a, hue = 'smoker')
```

```python
# style (style emphasizes the distinction between categories through marker style)

import seaborn as sns

a = sns.load_dataset('tips').tail(10)
print(a)

sns.relplot( x = 'total_bill', y = 'tip', data = a, style = 'sex')
```

```python
# size (size emphasizes the distinction between categories through size)

import seaborn as sns

a = sns.load_dataset('tips').sample(10)
print(a)

sns.relplot( x = 'tip', y = 'total_bill', data = a, size = 'day')
```

```python
# using hue,style & size together

import seaborn as sns

a = sns.load_dataset('tips').sample(10)
print(a)

sns.relplot( x = 'tip', y = 'total_bill', data = a, hue = 'day', style = 'size', size = 'smoker')
```
