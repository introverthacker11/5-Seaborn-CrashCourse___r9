```python
# box plot

import pandas as pd
import seaborn as sns

a = pd.read_csv('groupby(csvfile8).csv')
sns.catplot( x = 'College', y = 'Height(foot.inch)', data = a, kind = 'swarm')
sns.catplot( x = 'College', y = 'Height(foot.inch)', data = a, kind = 'box')

# 1st Quartile (Q1): This is the value below which 25% of the data fall. It is the median of the lower half of the dataset.
# Median (Q2): This is the value that separates the higher half from the lower half of the data. It's the middle value when the dataset is ordered.
# 3rd Quartile (Q3): This is the value below which 75% of the data fall. It is the median of the upper half of the dataset.

# Important Points:

# 1) If there are two minimum values then it will not show the minimum's value's line, the Q1 will be shown there.
# 2) If maximum value is any special value, quartile be shown on it's place and maximum's value's line will be shown on 2nd highest value.
# 3) If there is only one value, only 1 line will be shown.
```

```python
# hue

import pandas as pd
import seaborn as sns

a = pd.read_csv('groupby(csvfile8).csv')
b = sns.load_dataset('tips')
sns.catplot( x = 'College', y = 'Age', data = a, hue = 'Department')
sns.catplot( x = 'College', y = 'Age', data = a, hue = 'Department', kind = 'box')
sns.catplot( x = 'day', y = 'total_bill', data = b, hue = 'sex', kind = 'box')
```

```python
# dodge (puts hue in single plot)

import seaborn as sns
import pandas as pd

a = sns.load_dataset('tips')
sns.catplot( x = 'day', y = 'total_bill', data = a, kind = 'box', hue = 'sex', dodge = False)
```
