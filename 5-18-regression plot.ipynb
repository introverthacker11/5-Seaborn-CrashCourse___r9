##
## regplot:
##

```python
# regplot ( it doesn't support hue, row, cols etc)

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

a = pd.read_csv('groupby(csvfile8).csv')
sns.regplot( x = 'Rank', y = 'Age', data = a)
plt.grid( color = 'k')
# the line and ci shows the probability that how the age will increase or decrease with respect to rank
```

```python
# regplot ( it doesn't support hue, row, cols etc)

import seaborn as sns

a = sns.load_dataset('tips')
a.head()
sns.regplot( x = 'total_bill', y = 'tip', data = a)
sns.set_style('darkgrid')
# the line and ci shows the probability that how the tips will increase with respect to totalbill
```

## lmplot:
##

```python
# lmplot ( it support hue, col, row)

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

a = pd.read_csv('groupby(csvfile8).csv')
a,sns.lmplot( x = 'Rank', y = 'Age', hue = 'Department', data = a)
plt.grid()
```

```python
# lmplot

import seaborn as sns

a = sns.load_dataset('tips')
sns.lmplot( x = 'total_bill', y = 'tip', data = a, hue = 'smoker')
```
