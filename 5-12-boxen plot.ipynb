```python
# boxen plot

import seaborn as sns

a = sns.load_dataset('tips')
sns.catplot( x = 'total_bill', y = 'day', data = a, kind = 'swarm', height = 4, aspect = 1.3)
sns.catplot( x = 'total_bill', y = 'day', data = a, kind = 'boxen', hue = 'smoker', height = 4, aspect = 1.3)
sns.catplot( x = 'total_bill', y = 'day', data = a, kind = 'boxen', height = 4, aspect = 1.3)

```

```python
# hue

import seaborn as sns
import matplotlib.pyplot as plt 

a = sns.load_dataset('diamonds')
sns.catplot( x = 'color', y = 'price', kind = 'boxen', data = a, hue = 'cut') # the width shows number of data
```

```python
# dodge

import seaborn as sns
import matplotlib.pyplot as plt 

a = sns.load_dataset('diamonds')
sns.catplot( x = 'color', y = 'price', kind = 'boxen', data = a, hue = 'cut', dodge = False) # the width shows number of data
```
