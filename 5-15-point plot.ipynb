```python
# point plot

import seaborn as sns

a = sns.load_dataset('tips')
sns.catplot( x = 'day', y = 'tip', data = a, hue = 'smoker', height = 3, aspect = 1.3)
sns.catplot( x = 'day', y = 'tip', data = a, hue = 'smoker', kind = 'point', height = 3, aspect = 1.3)
```

```python
# estimator

import seaborn as sns
import matplotlib.pyplot as plt

a = sns.load_dataset('tips')
sns.catplot( x = 'day', y = 'tip', data = a, hue = 'smoker', kind = 'point', height = 3, aspect = 1.3)
plt.grid()
sns.catplot( x = 'day', y = 'tip', data = a, hue = 'smoker', kind = 'point', height = 3, aspect = 1.3, estimator = 'max')
plt.grid()
sns.catplot( x = 'day', y = 'tip', data = a, hue = 'smoker', kind = 'point', height = 3, aspect = 1.3, estimator = 'min')
plt.grid()
sns.catplot( x = 'day', y = 'tip', data = a, hue = 'smoker', kind = 'point', height = 3, aspect = 1.3, estimator = 'mean')
plt.grid()
sns.catplot( x = 'day', y = 'tip', data = a, hue = 'smoker', kind = 'point', height = 3, aspect = 1.3, estimator = 'median')
plt.grid()
```

```python
# estimator and grid through loop

import seaborn as sns
import matplotlib.pyplot as plt

a = sns.load_dataset('tips')

kind = ['point','point','point','point']
estim = ['max','min','mean','median']

for i in range(4):
    sns.catplot( x = 'day', y = 'tip', data = a, hue = 'smoker', kind = kind[i], estimator = estim[i], height = 3, aspect = 1.3)
    plt.grid()
plt.show()
```
