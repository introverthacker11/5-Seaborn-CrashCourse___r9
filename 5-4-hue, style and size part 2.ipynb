```python
# sizes (setting size limit in size)

import seaborn as sns

a = sns.load_dataset('tips').sample(10)

sns.relplot( x = 'tip', y = 'total_bill', data = a, size = 'smoker', sizes = (15,250))
```

```python
# palette (setting colors in hue)

import seaborn as sns
import pandas as pd

a = pd.read_csv('groupby(csvfile8).csv').sample(10)

colors = { 'Pre-Eng' : 'Red' , 'Comp.Science' : 'Yellow' , 'Pre-Medical' : 'Blue'}
sns.relplot( x = 'College', y = 'Rank', data = a, hue = 'Department', palette = colors)
```

```python
# markers (setting style in style)

import pandas as pd
import seaborn as sns

a = pd.read_csv('groupby(csvfile8).csv')

styles = { 19 : 'o' , 20 : 'h' , 21 : 'v' , 23 : 's'}
sns.relplot( x = 'College', y = 'Department', style = 'Age', data = a, markers = styles)
```

```python
# using style, markers & sizes together

import seaborn as sns

a = sns.load_dataset('tips')

b = { 'Female' : 'Red' , 'Male' : 'Blue'}
c = { 'No' : 'o' , 'Yes' : 'v'}

sns.relplot( x = 'total_bill', y = 'tip', data = a, hue = 'sex', palette = b, style = 'smoker', markers = c, size = 'time', sizes = (15,250))
```

```python
# col

import seaborn as sns

a = sns.load_dataset('tips')

b = { 'Female' : 'Red' , 'Male' : 'Blue'}
c = { 'No' : 'o' , 'Yes' : 'v'}

sns.relplot( x = 'total_bill', y = 'tip', data = a, hue = 'sex', palette = b, style = 'smoker', markers = c, size = 'time', sizes = (15,250), col = 'smoker', col_wrap = 2)
```
