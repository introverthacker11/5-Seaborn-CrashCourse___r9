```python
# bar plot (bar plot tells the average of all values)

import seaborn as sns

a = sns.load_dataset('tips')
print(a)
sns.catplot( x = 'day', y = 'total_bill', data = a, kind = 'bar')
sns.catplot( x = 'day', y = 'total_bill', data = a, kind = 'swarm')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

a = pd.read_csv('groupby(csvfile8).csv')
sns.catplot( x = 'College', y = 'Age', data = a, kind = 'swarm', aspect = 1.2, height = 3)
sns.catplot( x = 'College', y = 'Age', data = a, kind = 'bar', aspect = 1.2, height = 3)
plt.grid(color = 'k')

# here bar plot of caa is coming on 19.75 because (19+19+20+21)/4 = 19.75, i.e: average is 19.75
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

a = pd.read_csv('groupby(csvfile8).csv')
sns.catplot( x = 'College', y = 'Age', data = a, kind = 'swarm', aspect = 1.2, height = 3, hue = 'Department')
sns.catplot( x = 'College', y = 'Age', data = a, kind = 'bar', aspect = 1.2, height = 3)
plt.grid(color = 'k')
sns.catplot( x = 'College', y = 'Age', data = a, kind = 'bar', aspect = 1.2, height = 3, hue = 'Department') # it will tell the average according to each value of hue
plt.grid(color = 'k')

# here bar plot of caa is coming on 19.75 because (19+19+20+21)/4 = 19.75, i.e: average is 19.75
```
