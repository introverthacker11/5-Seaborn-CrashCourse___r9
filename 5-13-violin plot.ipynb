```python
# violin     (width of violin says that more data is around these datapoints)

import seaborn as sns
import matplotlib.pyplot as plt

a = sns.load_dataset('tips')
sns.catplot( x = 'total_bill', y = 'smoker', data = a)
plt.grid( color = 'k')
sns.catplot( x = 'total_bill', y = 'smoker', data = a, kind = 'violin')
plt.grid( color = 'k')
```

```python
# split

import seaborn as sns

a = sns.load_dataset('tips')
sns.catplot( x = 'total_bill', y = 'day', data = a, kind = 'violin', hue = 'sex', split = True)
plt.grid( color = 'k')
```

```python
# violin + swarm

import seaborn as sns

a = sns.load_dataset('tips')
b = sns.catplot( x = 'day', y = 'total_bill', data = a, kind = 'violin', hue = 'smoker', split = True, color= 'c')
sns.swarmplot( x = 'day', y = 'total_bill', data = a, ax = b.ax, color = 'k')
plt.grid( color = 'k')
```
