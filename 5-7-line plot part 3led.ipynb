```python
# cube_helix

import seaborn as sns

a = sns.load_dataset('dots')
b = sns.cubehelix_palette(light = 0.5, n_colors = 6)
a,sns.relplot( x = 'time', y = 'firing_rate', data = a, kind = 'line', hue = 'coherence', palette = b, style = 'choice')
```
